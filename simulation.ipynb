```python
import numpy as np
import numpy.random as npr
import pandas as pd
import time 
from random import * 
```

```python
class Plat : 
    def __init__(self ,ide , nom , pmin , pmax , stock):
        self.id = ide 
        self.nom = nom 
        self.pmin = pmin
        self.pmax = pmax 
        self.stock = stock 
        pas = (pmax - pmin) / (k - 1)
        self.liste_prix = [(pmin + i * pas) for i in range(k)]
    
    def afficher(self):
        print("id : " , self.id , "\nnom: " , self.nom , "\n(pmin , pmax) : (", self.pmin , "  , " , self.pmax , ")\nstock : " , self.stock )
```

```python
#Creation p plats 
Np = 10 
k = 3
list_plats = []
for i in range(Np):
    nom = "plat_" + str(i+1)
    ide = i+1 
    pmin = npr.randint(1 , 5)
    pmax = npr.randint(10 , 15)  
    stock = npr.randint(80 , 121)
    list_plats.append(Plat(ide , nom , pmin , pmax , stock))
for pl in list_plats : 
    pl.afficher()
```

```python
DEB = time.time()

#Df = pd.DataFrame(columns = [ "Client" , "Classa" ,"Classf" ,"Mois" ,"Jour" , "Periode" , "Plat" , "Note" , "Prix" , "Achat" ])

fichier = open("res2.txt" , "w")
fichier.write('"Mois","Jour","Periode","Client","Classf","Classa","Propension","Plat","Note","Prix","Achat"'+'\n')
N_utilisateurs = 20000
l_clients = range(N_utilisateurs)

Nm = 10 
Nj = 30
nbre_etats = 24 


Class_f  = {"pauvre" : 1/4 , "moyen" : 1/2 , "riche" : 1/4}
Class_a  = {"aime_pas" : 1/4 , "indifferent" : 1/2 , "aime" : 1/4}


Class_j = {"debut":list(range(3 , 10)) , "milieu" : list(range(10 , 25)) , "fin" :[1 , 2] + list(range(25 , Nj+1))}
Class_p = {"debut":list(range(6)) , "milieu" : list(range(6 , 18)) , "fin" :list(range(18 , nbre_etats+1))}


prop_seg_f = npr.multinomial(N_utilisateurs , list(Class_f.values())) 
liste_prop_seg_f = []
temp = l_clients
for i in range(len(prop_seg_f)):
    liste_prop_seg_f.append(sample(temp ,prop_seg_f[i] ))
    temp = [g for g in temp if g not in liste_prop_seg_f[i]]


lplnom = []
for plat in list_plats :
    lplnom.append(plat.nom)
    
Class_j_vals = []
Class = [Class_f , Class_a]

for mois in range( 1 , Nm+1):   
    
    
    deb = time.time() 
    print("debut du mois : " , mois)
    lamda = npr.uniform(30, 50) 
    prix_pauvre_m = npr.uniform(5, 10) 
    prix_moyen_m = npr.uniform(1.5, 2) * prix_pauvre_m 
    prix_riche_m = npr.uniform(2, 3) * prix_moyen_m 
    Class_prix_m = [prix_pauvre_m , prix_moyen_m ,  prix_riche_m]
    
    
    
    
    for jour in range(1 , Nj+1): 
        
        
        
        print("Debut du jour : " , jour)
        lamda2 = npr.poisson(lamda) 
        Class_prix_j =[]        
        for cpt in range(len(Class_prix_m)):            
            Class_prix_j.append(npr.poisson(Class_prix_m[cpt])) 
            
        
        if (jour>=3 and jour<10): 
            Class_prix_j = [npr.uniform(1 , 1.5)*temp for temp in Class_prix_j]
            
       
        if (jour>=25 or jour <3): 
            Class_prix_j = [npr.uniform(1 , 2)*temp for temp in Class_prix_j]
                        
        liste_clients_r =  l_clients 

        
        for periode in range(nbre_etats): 
            
            Class_prix_p =[] 
            for cpt in range(len(Class_prix_j)):            
                Class_prix_p.append(npr.poisson(Class_prix_j[cpt]))
                
            if periode in Class_p["milieu"]: 
                Class_prix_p = [npr.uniform(1 , 2)*temp for temp in Class_prix_p]
                
            if periode in Class_p["fin"]: 
                Class_prix_p = [npr.uniform(1 , 1.5)*temp for temp in Class_prix_p]
                
            
            Dic_prix = {}
            for plat in list_plats : 
                Dic_prix[plat] = npr.choice(plat.liste_prix)
                
                                
                
            N = npr.poisson(lamda2) 
            liste_clients_p = sample(liste_clients_r ,N ) 
            liste_clients_r = [i for i in liste_clients_r if not i in liste_clients_p] 
            
            for client in liste_clients_p: 
                
               
                
                ind_f = 0 
                for ind in range (len(Class_f)):
                    
                    if client in liste_prop_seg_f[ind]:
                        ind_f = ind 
                        
                Prix = list(Dic_prix.values()) 
                temp_a = []
                Dic_proppay = [] 
                propmax = 0 
                Note = [] 
                Indmax = []
                for ind in range(Np):
                     
                    propension = Class_prix_p[ind_f]
                    prix = Prix[ind]
                    cpt_a = choices(range(len(Class_a)) , list(Class_a.values())) 
                    cpt_a = cpt_a[0]
                    if cpt_a == 0 : 
                        note = npr.randint(1 , 4)
                    if cpt_a ==1 :  
                        note = npr.randint(1 , 6)
                    if cpt_a ==2 : 
                        note = npr.randint(3 , 6) 
                    Note.append(note)
                    propension = propension*(note/3)
                    Dic_proppay.append(propension)
                    
                    temp_a.append(list(Class_a.keys())[cpt_a])
                    #print(ind , " : ", " propension : ", propension ,"prix : ", prix , " rapport : " , (propension / prix))
                    if (propmax < (propension / prix) ) :
                        Indmax = []
                        propmax = propension / prix                         
                        indmax = ind 
                        ab=0 

                    if (propmax == (propension / prix) ) :  # si egalitÃ©                         
                        Indmax.append(ind) ; 
                        ab=1 
                Achat = [np.NaN for i in range(Np)]
                
                #print(indmax , " Et " , Indmax)
                if (propmax >= 1): 
                    if(ab==0):
                        Achat[indmax] = 1
                    if(ab==1):
                        Achat[choice(Indmax)] = 1 
               
                temp1 = list(Class_f.keys())[ind_f]
                for oo in range(Np):
                    fichier.write(str(mois)+","+str(jour)+","+str(periode)+","+str(client)+","+str(temp1)+","+str(temp_a[oo])+","+str(Dic_proppay[oo])+","+str(lplnom[oo]) +","+str(Note[oo]) +","+str(Prix[oo]) +","+str(Achat[oo]) + "\n")
    
                
            
    print(time.time() - deb)     
print(time.time() - DEB)            
fichier.close()            
           
```

```python

data = aaa[["Jour" , "Periode" , "Prix"]]
targets = aaa["Achat"]
print(data.mean() , " -- ", data.std())
from tensorflow import keras as k 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#data = scaler.fit_transform(data)
model = k.models.Sequential()
#24p 30jrs 3 prix 
model.add(k.layers.Dense(5 , activation ="relu"))
model.add(k.layers.Dense(10 , activation ="relu"))
model.add(k.layers.Dense(10 , activation ="relu"))
model.add(k.layers.Dense(1 , activation  = "linear"))
#print(data.mean() , " -- ", data.std())
```

```python
model.compile(optimizer='rmsprop',
              loss='mae')
data , targets = np.array(data) , np.array(targets)
model.fit(data , targets , epochs=30)
```
